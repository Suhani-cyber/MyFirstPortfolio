# Gemini with Google AI Studio(Gemini API)

```python
from google.colab import userdata

import os
from google import genai
from google.genai import types

class GeminiConversation:
    def __init__(self, api_key: str):
        self.client = genai.Client(api_key=api_key)
        self.history = []

    def generate_conversation_response(self, model: str, prompt: str, mime_type: str = "text/plain") -> str:
        # Add user's new prompt to conversation
        self.history.append(
            types.Content(
                role="user",
                parts=[types.Part.from_text(text=prompt)],
            )
        )

        # Create configuration
        generate_content_config = types.GenerateContentConfig(response_mime_type=mime_type)

        # Stream the response from the model
        response_text = ""
        for chunk in self.client.models.generate_content_stream(
            model=model,
            contents=self.history,
            config=generate_content_config,
        ):
            response_text += chunk.text

        # Save the model's response to history
        self.history.append(
            types.Content(
                role="model",
                parts=[types.Part.from_text(text=response_text)],
            )
        )

        return response_text

if __name__ == "__main__":
    # Example usage
    gemini_api_key = userdata.get("GEMINI_KEY")
    convo = GeminiConversation(api_key=gemini_api_key)

    response_1 = convo.generate_conversation_response(
        model="gemini-2.0-flash",
        prompt="What is the next number in this sequence: 2, 4, 8, 16, __?",
        mime_type="text/plain"
    )
    print("Response 1:", response_1)


```

```python
from google import genai
from google.genai import types
from IPython.display import Markdown, display

def display_markdown(text):
  display(Markdown(text))

class VertexGeminiConversation:
    def __init__(self, project: str, location: str):
        self.client = genai.Client(vertexai=True, project=project, location=location)
        self.history = []

    def generate_conversation_response(
        self,
        model: str,
        prompt: str,
        temperature: float = 1.0,
        top_p: float = 0.95,
        max_tokens: int = 8192,
        response_modalities: list = ["TEXT"],
        safety_settings: list = None,
        tools = None,
    ) -> str:
        # Add user's prompt
        self.history.append(
            types.Content(
                role="user",
                parts=[types.Part.from_text(text=prompt)],
            )
        )

        # Default safety settings if none provided
        if safety_settings is None:
            safety_settings = [
                types.SafetySetting(category="HARM_CATEGORY_HATE_SPEECH", threshold="OFF"),
                types.SafetySetting(category="HARM_CATEGORY_DANGEROUS_CONTENT", threshold="OFF"),
                types.SafetySetting(category="HARM_CATEGORY_SEXUALLY_EXPLICIT", threshold="OFF"),
                types.SafetySetting(category="HARM_CATEGORY_HARASSMENT", threshold="OFF"),
            ]

        # Configuration
        config = types.GenerateContentConfig(
            temperature=temperature,
            top_p=top_p,
            max_output_tokens=max_tokens,
            response_modalities=response_modalities,
            safety_settings=safety_settings,
            tools = tools,
        )

        # Generate model output
        response_text = ""
        for chunk in self.client.models.generate_content_stream(
            model=model,
            contents=self.history,
            config=config,
        ):
            response_text += chunk.text

        # Save model response
        self.history.append(
            types.Content(
                role="model",
                parts=[types.Part.from_text(text=response_text)],
            )
        )

        return response_text


# Example usage

vertex_convo = VertexGeminiConversation(
    project=userdata.get("GCP_Project_id"),
    location="us-central1"
)

response_1 = vertex_convo.generate_conversation_response(
    model="gemini-2.0-flash-001",
    prompt="Who is president of USA?"
)
display_markdown(response_1)
```

```python
response_2 = vertex_convo.generate_conversation_response(
    model="gemini-2.0-flash-001",
    prompt="Tell me more about him?"
)
display_markdown(response_2)
```

```python
from google.genai.types import (
    GenerateContentConfig,
    GenerateContentResponse,
    Part,
    Tool
)


def print_grounding_data(response: GenerateContentResponse) -> None:
    candidate = response.candidates[0]

    citation_metadata = getattr(candidate, "citation_metadata")
    if citation_metadata:
        print("Citations")
        for citation in citation_metadata.citations:
            print(citation)

    grounding_metadata = getattr(candidate, "grounding_metadata")
    if grounding_metadata:
        print("\nGrounding Chunks")
        for grounding_chunk in grounding_metadata.grounding_chunks:
            print(grounding_chunk)

        print("\nGrounding Supports")
        for grounding_support in grounding_metadata.grounding_supports:
            print(grounding_support)


PROMPT = "What is the current temperature at this location?"

vertex_client = genai.Client(vertexai=True, project=userdata.get("GCP_Project_id"), location="us-central1")

response = vertex_client.models.generate_content(
    model="gemini-2.5-pro-exp-03-25",
    contents=[
        Part.from_uri(
            file_uri="gs://github-repo/generative-ai/gemini/grounding/paris.jpg",
            mime_type="image/jpeg",
        ),
        PROMPT,
    ],
    config=GenerateContentConfig(
        tools=[types.Tool(google_search=types.GoogleSearch())],
    ),
)

display(Markdown(response.text))
print("------------------")
print_grounding_data(response)
```

```python
generation_model = "imagen-3.0-generate-002"
generation_model_fast = "imagen-3.0-fast-generate-001"
```

```python
import typing

import IPython.display
from PIL import Image as PIL_Image
from PIL import ImageOps as PIL_ImageOps


def display_image(
    image,
    max_width: int = 600,
    max_height: int = 350,
) -> None:
    pil_image = typing.cast(PIL_Image.Image, image._pil_image)
    if pil_image.mode != "RGB":
        # RGB is supported by all Jupyter environments (e.g. RGBA is not yet)
        pil_image = pil_image.convert("RGB")
    image_width, image_height = pil_image.size
    if max_width < image_width or max_height < image_height:
        # Resize to display a smaller notebook image
        pil_image = PIL_ImageOps.contain(pil_image, (max_width, max_height))
    IPython.display.display(pil_image)
```

```python
import matplotlib.pyplot as plt

prompt = """
# A photo of a modern building with water in the background.

"""

# Imagen 3 image generation
image = vertex_client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="3:4",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="DONT_ALLOW",
    ),
)

# Imagen 3 Fast image generation
fast_image = vertex_client.models.generate_images(
    model=generation_model_fast,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="3:4",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="DONT_ALLOW",
    ),
)

# Display generated images
fig, axis = plt.subplots(1, 2, figsize=(12, 6))
axis[0].imshow(image.generated_images[0].image._pil_image)
axis[0].set_title("Imagen 3")
axis[1].imshow(fast_image.generated_images[0].image._pil_image)
axis[1].set_title("Imagen 3 Fast")
for ax in axis:
    ax.axis("off")
plt.show()
```

```python
from google.genai.types import (
    FunctionDeclaration,
    Part,
    Schema,
    Tool,
    ToolConfig,
    Type,
)
```

```python
get_usd_to_inr_rate = FunctionDeclaration(
      name= "get_usd_to_inr_rate",
      description= "Fetches the current USD to INR exchange rate.",
      parameters= {
        "type": "object",
        "properties": {},
        "required": []
      }
    )


get_weather_forecast_meteostat = FunctionDeclaration(
    name= "get_weather_forecast_meteostat",
    description= "Fetches the weather forecast for the next 10 days.",
    parameters= {
      "type": "object",
      "properties": {},
      "required": []
    }
)

get_today_date = FunctionDeclaration(
    name= "get_today_date",
    description= "Fetches the current date.",
    parameters= {
      "type": "object",
      "properties": {},
      "required": []
    }
)


get_time = FunctionDeclaration(
    name= "get_time",
    description= "Fetches the current time.",
    parameters= {
      "type": "object",
      "properties": {},
      "required": []
    }
)


all_function_tool = Tool(
    function_declarations=[
        get_usd_to_inr_rate,
        get_weather_forecast_meteostat,
        get_today_date,
        get_time,
    ],
)


# config = GenerateContentConfig(temperature=0, tools=[all_function_tool])
# config.tool_config = ToolConfig(
#     function_calling_config=FunctionCallingConfig(
#         mode=FunctionCallingConfigMode.AUTO,  # The default model behavior. The model decides whether to predict a function call or a natural language response.
#     )
# )
```

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from meteostat import Point, Daily
from currency_converter import CurrencyConverter
cc_object = CurrencyConverter()

def get_usd_to_inr_rate_func():
  print("(py function) fetching the USD to INR Rate")
  ans = cc_object.convert(1, 'USD', 'INR')
  ans = round(ans, 4)
  return ans

def get_weather_forecast_meteostat_func():
    print("(py function) fetching the weather data for next 10 days")
    city = "Ahmedabad"
    lat = 23.0225
    lon = 72.5714

    start = datetime.now()
    end = start + timedelta(days=10)

    # Create Point for the location
    location = Point(lat, lon)

    # Get daily data for the location between start and end dates
    data = Daily(location, start, end)
    data = data.fetch()

    # Extract dates and temperatures
    dates = data.index.strftime('%Y-%m-%d')
    temperatures = data['tavg']  # Use 'tmax' for maximum temperature or 'tavg' for average

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(dates, temperatures, marker='o', linestyle='-')
    plt.xlabel("Date")
    plt.ylabel("Temperature (°C)")
    plt.title(f"Avg Weather Forecast for {city} (Next 10 Days)")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def get_today_date_func():
    print("(py function) fetching the current date")
    return datetime.today().strftime("%Y-%m-%d")


def get_time_func():
    print("(py function) fetching the current time")
    return datetime.today().strftime("%H:%M:%S")
```

```python
function_handler = {
    "get_usd_to_inr_rate": get_usd_to_inr_rate_func,
    "get_weather_forecast_meteostat": get_weather_forecast_meteostat_func,
    "get_today_date": get_today_date_func,
    "get_time": get_time_func,
}
```
